# 皮马印第安人糖尿病数据可视化

## 1.0 读取数据

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
#使绘图内嵌到 notebook 中
%matplotlib inline

data=pd.read_csv('dataset/pima.csv',header=None,
                 names=['pregnant','concentration',
                        'Diastolic blood pressure',
                        'Triceps skin fold thickness',
                        '2-Hour serum insulin',
                        'Body mass index',
                        'Diabetes pedigree function',
                        'Age','Class variable'])
data.head()
```

## （1）任选两个字段绘制散点图。

```python
ax = data[data['Class variable']==1].plot(x='Diabetes pedigree function', y='concentration',
               kind='scatter', c='red', ax=None,alpha=0.7,
               label='Class variable:1')
data[data['Class variable']==0].plot(x='Diabetes pedigree function', y='concentration',
               kind='scatter', c='blue', ax=ax,alpha=0.7,
               label='Class variable:2')
```

## （2）使用全部或者部分特征绘制散布图。

```python
color = {1:'red', 0:'blue'}
pd.plotting.scatter_matrix(data.iloc[:,:-1], figsize=(16, 16), diagonal='kde', marker='o',
                  s=40, alpha=0.4, c=data['Class variable'].apply(lambda x: color[x]))
```

## （3）绘制调和曲线图。

```python
from pandas.plotting import andrews_curves
andrews_curves(data, 'Class variable',color=['red','blue'])
```
